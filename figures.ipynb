```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 12})
plt.rcParams['font.family'] = 'Arial'

# Data
moe_training = [44.27, 51.49, 54.36, 57.75, 61.19]
baseline_training = [42.52, 51.14, 55.99, 58.62, 60.52]

# plt.figure(figsize=(4, 4))

# Plot both points and lines
plt.plot(moe_training, marker='o', color='b', label='MoE-ViT')
plt.plot(baseline_training, marker='o', color='r', label='Baseline-ViT')

# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')

# Set y limits between 0 and 100
plt.ylim(0, 100)

# Force X ticks to be integers
plt.xticks(range(5))

# Add a legend 
plt.legend()

# Save fig 
plt.savefig('fig2a.svg', dpi=600)
```
